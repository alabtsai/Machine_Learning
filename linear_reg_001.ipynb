## use Stanford cs20 birth rate and placeholder as a linear  regression example

```python
import tensorflow as tf
```

```python
import utils
```

```python
DATA_FILE = 'data/birth_life_2010.txt'
```

```python
data, n_samples = utils.read_birth_life_data(DATA_FILE)
```

```python
Xi = tf.placeholder(tf.float32, name='X')
Yo = tf.placeholder(tf.float32, name='Y')
```

```python
w = tf.get_variable('weights', initializer=tf.constant(0.0))
b = tf.get_variable('bias', initializer=tf.constant(0.0))
```

```python
Y_predicted = w * Xi + b
```

```python
loss = tf.square(Yo - Y_predicted, name='loss')
optimizer = tf.train.GradientDescentOptimizer(learning_rate=0.001).minimize(loss)
```

```python
writer = tf.summary.FileWriter('./graphs/linear_reg', tf.get_default_graph())
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer()) 
    #  train the model for 20 epochs
    for i in range(100):
        total_loss = 0
        count=0
        for xq,yq in data:
            #print(xq,yq)
            _, l = sess.run([optimizer, loss], feed_dict={Xi: xq, Yo:yq})
            total_loss += l
            count=count+1
        w_out, b_out = sess.run([w, b])
        print('Epoch {0}: {1}'.format(i, total_loss/n_samples))
        print("count=",count)
        print(w_out, b_out)
        print(" ")
    writer.close() 
    w_out, b_out = sess.run([w, b])
```
